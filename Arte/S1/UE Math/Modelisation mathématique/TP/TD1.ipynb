```python
def bisection_method(f, a, b, e):
    cmp=0
    if b <= a or f(a)*f(b) >= 0:
        raise ValueError("Hypotse de base non vrifie")
    while(b-a>e):
        cmp+=1
        m = (a+b)/2
        y = f(m)
        if y == 0:
            return m, cmp
        elif (y*f(a))>0:
            a=m
        else:
            b=m
    return ((a+b)/2, cmp)
```
