```python
def bisection_method(f, a, b, e):
    cmp=0
    if b <= a or f(a)*f(b) >= 0:
        raise ValueError("Hypotse de base non vrifie")
    while(b-a>e):
        cmp+=1
        m = (a+b)/2
        y = f(m)
        if y == 0:
            return m, cmp
        elif (y*f(a))>0:
            a=m
        else:
            b=m
    return ((a+b)/2, cmp)
```

```python
norm2(x, y)=sqrt((x-1)^2 + (y-1)^2 + (x+y-1)^2)
```

```python
x,y=var('x,y')
plot3d(norm2(x,y), (x,0,1),(y,0,1))
```

```python
contour_plot(norm2(x,y), (x,0,1),(y,0,1), cmap='jet',colorbar=True, contours=100)
```

```python
def compute_matrix(n):
    mat = []
    for i in range(n+1):
        mat.append([])
        for j in range(n+1):
            mat[i].append(norm2(i/n, j/n))
    return matrix(mat)
```

```python
compute_matrix(100).plot(cmap='jet')
```
